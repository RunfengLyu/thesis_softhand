```python
from stable_baselines3 import PPO
# Create the environment
import softhand_sim
env = gymnasium.make('softhand_sim/SofthandTest-v0')  # Replace with your environment class

print("model included")
# Create the agent
model = PPO("MlpPolicy", env, verbose=1)

#HandEnv.reset = lambda self, **kwargs: self.reset(**kwargs)
# Train the agent
model.learn(total_timesteps=100)

# Save the trained agent
model.save("ppo_agent")

# Test the trained agent
obs,info = env.reset()
print(info["frames"])
#media.show_video(info["frames"])

print("test_obs", obs)
obs = np.array(obs).reshape((-1,1)).squeeze()
for i in range(1000):
    action, _states = model.predict(obs, deterministic=True)
    obs, reward, done, truncated, info = env.step(action)
    if reward:
      obs = env.reset()
```
